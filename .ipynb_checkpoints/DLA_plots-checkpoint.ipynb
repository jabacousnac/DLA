```python
from matplotlib import pyplot as plt
import numpy as np

fig, ax = plt.subplots()
ax.set_xlabel(r'ln(R)')
ax.set_ylabel(r'ln(N)')
X = [2.302585093, 2.708050201, 1.609437912, 0.693147181, 3.218875825, 2.995732274, 3.496507561, 2.079441542, 2.890371758]
Y = [4.94875989, 5.545177444, 3.465735903, 1.609437912, 6.349138991, 5.945420609, 6.665683718, 4.290459441, 5.717027701]
ax.scatter(X,Y,c='red')
p = np.polyfit(X,Y,deg=1)
p0, p1 = np.around(p[0],decimals=2), np.around(p[1],decimals=2)
X_hat = np.linspace(0,3.5,50)
ax.plot(X_hat, X_hat*p[0]+p[1],'b-',label='{}X + {}'.format(p0,p1))
ax.legend()
plt.savefig('fractal_dimension.png', dpi = 200)
```
